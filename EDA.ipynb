# Importing Libraries and Loading Dataset



```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
df=pd.read_csv('WineQT.csv')
```

# Histogram

```python
plt.figure(figsize=(5, 4))
plt.hist(df['alcohol'], bins=10, color='skyblue', edgecolor='black')
plt.title('Distribution of Alcohol Content')
plt.xlabel('Alcohol')
plt.ylabel('Frequency')
plt.show()
```

# Bar Graph

```python
quality_counts = df['quality'].value_counts()
plt.figure(figsize=(5, 4))
sns.barplot(x=quality_counts.index, y=quality_counts.values, palette="viridis")
plt.title('Quality Count')
plt.xlabel('Quality(Out of 10)')
plt.ylabel('Count of Samples')
plt.show()
```

# PieChart

```python
plt.figure(figsize=(5, 5))
colors = sns.color_palette("Set2")
quality_counts = df['quality'].value_counts()
wedges, texts, autotexts = plt.pie(
    quality_counts, labels=None, autopct='%1.1f%%', colors=colors, startangle=140,
    wedgeprops={'edgecolor': 'black'}, textprops=dict(color="black")
)
plt.legend(wedges, quality_counts.index, title="Quality Levels", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))
plt.title('Quality Distribution')
plt.show()
```

## Box Plot

```python
plt.figure(figsize=(5, 4))
sns.boxplot(data=df[['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar']])
plt.title('Box Plot of Wine Attributes')
plt.show()
```

#Scatter Plot

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='alcohol', y='quality', data=df, hue='quality', palette='coolwarm')
plt.title('Alcohol vs Quality')
plt.xlabel('Alcohol')
plt.ylabel('Quality')
plt.show()
```

#Line Chart

```python
plt.figure(figsize=(5, 4))
sns.lineplot(data=df[['pH', 'sulphates', 'alcohol']])
plt.title('Line Plot of pH, Sulphates, and Alcohol')
plt.xlabel('Index')
plt.show()
```

# Violin Plot

```python
plt.figure(figsize=(5, 4))
sns.violinplot(x='quality', y='alcohol', data=df, palette='Set3')
plt.title('Violin Plot of Alcohol by Quality')
plt.show()
```

#Correlation Heatmap

```python
correlation_matrix=df.corr()
plt.figure(figsize=(5,4))
sns.heatmap(correlation_matrix,annot=True,cmap='coolwarm',fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```
